## Creating an SQLite Database

```python
import sqlite3 as sq3
```

```python
con = sq3.connect("movies.db")
```

```python
con.execute("Select * FROM sqlite_master").fetchall()
```

## Loading Data from DataFrames into an SQLite Database

```python
import pandas as pd
```

```python
data = pd.read_csv('movies_complete.csv')
```

```python
data
```

```python
Movies = data[["id", "title","genres", "revenue_musd", "budget_musd", "belongs_to_collection", "release_date"]].copy()
Movies
```

```python
Votes = data[["id", "vote_count", "vote_average"]].copy()
Votes
```

```python
Genres = data[["id", "genres"]].copy()
Genres
```

```python
Movies.to_sql("Movies", con, index = False)
```

```python
Votes.to_sql("Votes", con, index = False)
```

```python
Genres.to_sql("Genres", con, index = False)
```

```python
con.execute("Select * FROM sqlite_master").fetchall()
```

## Join Queries

```python
pd.read_sql("SELECT * \
            FROM Movies \
            JOIN Votes \
            ON Movies.id=Votes.id", con)
```

```python
pd.read_sql("SELECT Movies.id, Movies.title, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id=Votes.id \
            WHERE Votes.vote_average > 8", con, index_col = "id")
```

```python
pd.read_sql("SELECT Movies.id, Movies.title, Movies.budget_musd, Votes.vote_average \
            FROM Movies \
            JOIN Votes \
            ON Movies.id=Votes.id \
            WHERE Votes.vote_average > 8 \
            ORDER BY Movies.budget_musd ASC", con, index_col = "id")
```

```python
df2 = pd.read_sql("SELECT Genres.id, Genres.genres, Movies.revenue_musd, Movies.title \
            FROM Genres \
            LEFT JOIN Movies \
            ON Genres.id=Movies.id", con)
df2
```

```python
pd.read_sql("SELECT Genres.genres \
            FROM Genres \
            LEFT JOIN Movies \
            ON Genres.id=Movies.id \
            WHERE Movies.title = 'Frozen II'", con)
```

```python
pd.read_sql("SELECT Genres.genres \
            FROM Genres \
            LEFT JOIN Movies \
            ON Genres.id=Movies.id \
            WHERE Movies.title = 'Jumanji'", con)
```
